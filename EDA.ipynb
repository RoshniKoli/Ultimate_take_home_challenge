```python
import warnings
import itertools
import numpy as np
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")
plt.style.use('fivethirtyeight')
import pandas as pd
import statsmodels.api as sm
import matplotlib

matplotlib.rcParams['axes.labelsize'] = 14
matplotlib.rcParams['xtick.labelsize'] = 12
matplotlib.rcParams['ytick.labelsize'] = 12
matplotlib.rcParams['text.color'] = 'k'
```

```python
data = pd.read_json(r'C:\Users\roshn\Documents\DataAnalytics\Springboard\Take_home_challenge\ultimate_challenge\data\logins.json')
```

```python
data.info()
```

```python
data = data.sort_values(by='login_time')
```

```python
resampled = data.resample('15min',on='login_time').count()
```

```python

plt.style.use('fivethirtyeight')
_= plt.plot(resampled.login_time)
_= plt.tick_params(axis ='x', rotation = 45)
plt.rcParams["figure.figsize"] = (30,10)

plt.show()
```

```python
weekly = data.resample('W-MON',on = 'login_time').count()
```

```python
plt.style.use('fivethirtyeight')
_= plt.plot(weekly.login_time)
_= plt.tick_params(axis ='x', rotation = 45)

plt.show()
```

```python
monthly = data.resample('M',on = 'login_time').count()
plt.rcParams["figure.figsize"] = (12,8)
_= plt.plot(monthly.login_time)
_= plt.tick_params(axis ='x', rotation = 45)

plt.show()
```
